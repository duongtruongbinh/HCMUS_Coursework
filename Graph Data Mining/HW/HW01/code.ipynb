```python
import networkx as nx
import matplotlib.pyplot as plt

# Hm v mng Erdos-Renyi vi trung bnh bc k
def draw_graph(N, k):
    p = k / (N - 1)
    G = nx.erdos_renyi_graph(N, p)
    nx.draw(G, node_size=10)
    plt.title(f'Erdos-Renyi Network vi N={N}, <k>={k}')
    plt.show()

# Thng s ca mng
N = 500

# Trung bnh bc k
k_values = [0.8, 1, 8]

# V mng vi cc trung bnh bc khc nhau
for k in k_values:
    draw_graph(N, k)
```

```python
def randomEdgesGenerator(nodes: np.ndarray, p: float) -> np.array:
    """Hm pht sinh danh sch cnh ngu nhin

    Parameters
    ----------
    nodes
        (np.array): mt mng numpy cha cc ch s nh. V d: `[1, 2, 3, 4]`
    p
        (float): xc sut ca lin kt (c gi tr thc, chy t 0 ti 1).

    Returns
    -------
        (np.array) vi cc tuples c dng `[(1, 2), (2, 4)]`, trong  `1-2` l mt cnh nn c thm vo  th.
    """
    # YOUR CODE HERE
    edges = np.array([(i, j) for i in nodes for j in nodes if i < j and random.random() < p])
    return edges

def ERModelGenerator(n: int, p: float) -> nx.Graph:
    """Hm pht sinh  th ngu nhin da trn m hnh ErdsRnyi

    Parameters
    ----------
    n
        s nh ca  th
    p
        xc sut lin kt

    Returns
    -------
        (nx.Graph)  th ngu nhin
    """
    G = nx.Graph()
    nodes = np.arange(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(randomEdgesGenerator(nodes, p))
    return G

N = 3000
p = 0.0001
G = ERModelGenerator(N, p)
degree_seq = [degree for node, degree in G.degree]
bins, freq = np.unique(degree_seq, return_counts=True)
plt.bar(bins, freq)
```
