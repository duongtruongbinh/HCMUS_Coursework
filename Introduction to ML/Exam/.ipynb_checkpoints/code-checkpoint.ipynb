```python
def generate_data(num_samples):
    """
    Generates a dataset with attributes and labels based on given rules.

    Args:
        num_samples: Number of data points to generate.

    Returns:
        A list of dictionaries, where each dictionary represents a data point.
    """
    
    data = []
    # 80% data following rules
    
    for _ in range(int(0.8 * num_samples)):
    # Randomly choose a valid combination based on rules
        choice = random.choice([1, 2, 3, 4, 5])
        if choice == 1:
            data.append({
                "Gender": "Female",
                "Height": "Short",
                "Color_hair": "Black",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40, 150),
                "is_Asian": "Yes"
            })
        elif choice == 2:
            data.append({
                "Gender": "Female",
                "Height": "Tall",
                "Color_hair": "Blonde",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "No"
            })
        elif choice == 3:
            data.append({
                "Gender": "Male",
                "Height": "Medium",
                "Color_hair": "Black",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "Yes"
            })
        elif choice == 4:
            data.append({
                "Gender": "Male",
                "Height": "Medium",
                "Color_hair": "Blonde",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "No"
            })
        elif choice == 5:
            data.append({
                "Gender": "Male",
                "Height": "Short",
                "Color_hair": "Black",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "Yes"
            })
                

    # 20% wrong data
    for _ in range(int(0.2 * num_samples)):
        choice = random.choice([1, 2, 3, 4,5])
        if choice == 1:
            data.append({
                "Gender": "Female",
                "Height": "Tall",
                "Color_hair": "Blonde",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "Yes"
            })
        elif choice == 2:
            data.append({
                "Gender": "Female",
                "Height": "Short",
                "Color_hair": "Black",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "No"
            })
        elif choice == 3:
            data.append({
                "Gender": "Male",
                "Height": "Medium",
                "Color_hair": "Black",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "No"
            })
        elif choice == 4:
            data.append({
                "Gender": "Male",
                "Height": "Tall",
                "Color_hair": "Blonde",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "Yes"
            })
        elif choice == 5:
            data.append({
                "Gender": "Male",
                "Height": "Short",
                "Color_hair": "Black",
                "Color_eye": random.choice(["Black", "Brown", "Blue", "Gray"]),
                "Weight": random.uniform(40,150),
                "is_Asian": "No"
            })

    return data

# Example usage:
num_data_points = 1000
train_data = generate_data(num_data_points)
test_data = generate_data(num_data_points)
train_df = pd.DataFrame(train_data)
test_df = pd.DataFrame(test_data)
```

```python
from sklearn.tree import _tree

def tree_to_code(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    feature_names = [f.replace(" ", "_")[:-5] for f in feature_names]
    print("def predict({}):".format(", ".join(feature_names)))

    def recurse(node, depth):
        indent = "    " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print("{}if {} <= {}:".format(indent, name, np.round(threshold,2)))
            recurse(tree_.children_left[node], depth + 1)
            print("{}else:  # if {} > {}".format(indent, name, np.round(threshold,2)))
            recurse(tree_.children_right[node], depth + 1)
        else:
            print("{}return {}".format(indent, tree_.value[node]))

    recurse(0, 1)
    
tree_to_code(model, feature_names)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

def extract_rules(tree, feature_names, class_names):
    """
    Extracts rules from a decision tree, handling potential index errors.

    Args:
        tree: The trained decision tree model.
        feature_names: Names of the features used in the model.
        class_names: Names of the target classes.

    Returns:
        A list of tuples, where each tuple contains a rule (string) and its corresponding class name.
    """

    rules = []

    def recurse(node, rule, rules):
        if tree.children_left[node] != tree.children_right[node]:  # Check for non-leaf node
            try:  # Handle potential IndexError for missing features
                name = feature_names[tree.feature[node]]
                threshold = tree.threshold[node]
            except IndexError:
            # Handle missing features (e.g., not used in this split) by skipping the node
                return

            left_rule = rule + f" AND {name} <= {threshold:.2f}"
            recurse(tree.children_left[node], left_rule, rules)
            right_rule = rule + f" AND {name} > {threshold:.2f}"
            recurse(tree.children_right[node], right_rule, rules)
        else:
            class_idx = np.argmax(tree.value[node])  # Handle potential NaNs in value
            class_name = class_names[class_idx]
            rule = rule[5:]  # Remove leading " AND "
            rules.append((rule, class_name))

    recurse(0, "", rules)
    return rules

def calculate_support_confidence(rules, X_train_encoded, y_train):
    """
    Calculates support and confidence for each rule.

    Args:
        rules: A list of tuples containing rules and their corresponding class names.
        X_train_encoded: Encoded training features.
        y_train: Training target labels.

    Returns:
        A dictionary containing support values for each rule and a dictionary containing confidence values.
    """

    support = {}
    correct = {}

    for rule, class_name in rules:
        condition = eval(rule)  # Convert rule string to Python expression
        support_count = np.sum(np.all(X_train_encoded[:, condition], axis=1))
        correct_count = np.sum(np.all(X_train_encoded[:, condition], axis=1) & (y_train == class_name))
        support[rule] = support_count / len(X_train_encoded)
        correct[rule] = correct_count / support_count if support_count != 0 else 0

    return support, correct

def get_top_rules(support, correct, confidence_threshold=0.5):
    """
    Extracts the top 5 rules based on support and confidence, applying a confidence threshold.

    Args:
        support: A dictionary containing support values for each rule.
        correct: A dictionary containing confidence values for each rule.
        confidence_threshold: The minimum confidence level required for a rule to be considered (default: 0.5).

    Returns:
        A list of top 5 rules (tuples containing rule, support, and confidence values).
    """

    sorted_rules = sorted(support.items(), key=lambda x: x[1], reverse=True)
    top_rules = []

    for rule, sup in sorted_rules:
        if len(top_rules) >= 5:
            break
        if correct[rule] > confidence_threshold:
            top_rules.append((rule, sup, correct[rule]))

    return top_rules

# Example usage (assuming you have trained your model and encoded data)
rules = extract_rules(model.tree_, X_train.columns, ['Not Asian', 'Asian'])
support, confidence = calculate_support_confidence(rules, X_train_encoded, y_train)
top_rules = get_top_rules(support, confidence)

for i, (rule, sup, conf) in enumerate(top_rules, 1):
    print(f"Rule {i}: {rule} | Support: {sup:.2f} | Confidence: {conf:.2f}")
```

```python
from sklearn.tree import DecisionTreeClassifier

def extract_rules(tree, feature_names, target_names):
  """
  Hàm này trích xuất các luật từ cây quyết định sử dụng API công khai.

  Tham số:
      tree: Cây quyết định được đào tạo.
      feature_names: Tên các đặc trưng.
      target_names: Tên các mục tiêu.

  Trả về:
      Danh sách các luật.
  """

  # Lấy thông tin về các nút trong cây
  children_left = tree.children_left
  children_right = tree.children_right
  feature = tree.feature
  threshold = tree.threshold
  value = tree.value

  # Khởi tạo danh sách các luật
  rules = []

  # Duyệt qua các nút trong cây, bắt đầu từ nút gốc (node_id=0)
  def traverse(node_id):
    # Kiểm tra xem nút này có phải là nút lá hay không
    if children_left[node_id] == children_right[node_id]:
      # Tính toán confidence và support của luật
      class_idx = np.argmax(value[node_id])
      confidence = value[node_id, class_idx] / value[node_id].sum()
      support = (value[node_id].sum() / len(y_train))

      # Thêm luật vào danh sách
      rules.append({
          "antecedent": _get_antecedent(node_id, feature_names, threshold, children_left, children_right),
          "consequent": target_names[class_idx],
          "confidence": confidence,
          "support": support,
      })
    else:
      # Duyệt qua các con của nút
      traverse(children_left[node_id])
      traverse(children_right[node_id])

  traverse(0)  # Bắt đầu duyệt từ nút gốc

  # Sắp xếp các luật theo confidence và support
  rules.sort(key=lambda rule: (rule["confidence"], rule["support"]), reverse=True)

  return rules[:5]

def _get_antecedent(node_id, feature_names, threshold, children_left, children_right):
  """
  Hàm này lấy ra phần điều kiện của luật.

  Tham số:
      node_id: ID của nút trong cây.
      feature_names: Tên các đặc trưng.
      threshold: Ngưỡng của nút.
      children_left: Mảng con bên trái của nút.
      right_child: Mảng con bên phải của nút (sử dụng children_right cho tính nhất quán).

  Trả về:
      Phần điều kiện của luật.
  """

  if node_id == 0:
    return "True"

  feature_name = feature_names[feature[node_id]]
  threshold_value = threshold[node_id]

  return f"{feature_name} <= {threshold_value}" if children_left[node_id] else f"{feature_name} > {threshold_value}"

# Ví dụ sử dụng
# ...

target_names = ['Not Asian', 'Asian']

# Trích xuất các luật
rules = extract_rules(model.tree_, feature_names, target_names)

# In ra các luật
for rule in rules:
  print(f"Luật: {rule['antecedent']} => {rule['consequent']}")
  print(f"Confidence: {rule['confidence']}")
  print(f"Support: {rule['support']}")
  print()
```
