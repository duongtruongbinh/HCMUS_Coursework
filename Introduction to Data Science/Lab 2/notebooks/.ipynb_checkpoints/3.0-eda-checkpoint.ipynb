### Describe the correlation between features in the dataset (0.75 points)

Considering the available features within the training dataset, we want to identify and analyse the relationships between them, and then determine which features within the dataset significantly contribute to our solution goal. By using calculation and visualisation, Python allows us to create a correlation matrix, which is a table that represents the correlation coefficients between different variables.

Now let's make a correlation matrix, then visualize it, and describe what insights you observed?

```python
# PRE-PROCESSING IF NEEDED
# YOUR CODE HERE (OPTION) 
```

```python
# YOUR CODE HERE
# create a copy
df_copy = cleaned_df.copy()

# Exclude the categorical features from the matrix
df_copy.drop(['Year', 'Country'], inplace=True, axis='columns')

# plot a correlation matrix
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df_copy.corr(), cmap='RdBu', center=0,ax=ax)
plt.show()
```

Based on the correlation matrix, we can observe several features that are correlated with each other. We will ignore the correlations involving features such as `Total Population`, `Female Population`, and `Male Population` (as this is expected).

- `Female Employment in Agriculture(%)` and `Employment in Agriculture(%)`, as well as `Female Employment in Industry(%)` and `Employment in Industry(%)`, show a high positive correlation with each other but do not correlate significantly with `Female Population`. This may imply that the number of women in the population does not significantly affect the employment rates in agriculture and industry.

- `GDP in USD` and `Agriculture value added(in USD)` show a high positive correlation with `Total Population` (and, of course, with `Female Population` and `Male Population`). This means that countries with larger populations tend to have higher GDP and agriculture value added, which can be explained by several factors:
    - A larger population leads to a larger consumer market, increasing demand for goods and services, including agriculture.
    - A larger population provides a larger labor force, which helps countries develop their economies, including agriculture.
    - A larger population also leads to higher potential for innovation and creativity, contributing to economic growth in various sectors.

    However, there are exceptions (some countries with high GDP but relatively small populations, etc.). Using this correlation can help nations make more informed decisions about economic and population policies.

- `GDP in USD` and `Agriculture value added(in USD)` show a positive correlation with each other (because they are both correlated with `Total Population`), indicating that agriculture has a significant impact and plays an essential role in creating economic value.

- `Fossil Fuel Consumption (%)` has a negative correlation with `Electric Power Consumption(kWH per capita)` and `Renewable Energy Consumption (%)`. This means that countries that consume more fossil fuels tend to consume less electricity per capita and less renewable energy. This is an interesting relationship because fossil fuels are harmful to the environment, while electricity and renewable energy sources are cleaner and more sustainable.

- `Male life expectancy` and `Female life expectancy` have a high positive correlation because both are influenced by similar factors, such as healthcare conditions, quality of life, and socioeconomic conditions.

**Question 01**: How does the correlation between features change in the top 10 countries with the LARGEST population?

When we have an answer for this question, we might see how features interaction with each other when the conutry population is large.

```python
# PRE-PROCESSING IF NEEDED
# YOUR CODE HERE (OPTION)

# Get the top 10 contries with the largest population
largest_population_lst = cleaned_df.groupby('Country')['Total Population'].max().sort_values(ascending=False).head(10).index.tolist()
largest_population_df = cleaned_df[cleaned_df['Country'].isin(largest_population_lst)]
```

```python
# YOUR CODE HERE
largest_copy = largest_population_df.copy()

# Exclude the categorical features from the matrix
largest_copy.drop(['Year', 'Country'], inplace=True, axis='columns')

# plot a correlation matrix
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(largest_copy.corr(), cmap='RdBu', center=0,ax=ax)
plt.show()
```

**YOUR ANSWER**

- The overall correlation matrix for the top 10 most populous countries is higher and more distinct correlations (such as the correlations between features `Female Employment in Industry(%)`, `Employment in Industry(%)`, `Employment in Agriculture`, `Female Employment in Agriculture`...). This may be because this is a correlation matrix specific to these 10 countries, so the relationships between them are more accurately reflected.

- Some features have shifted from positive correlations to negative correlations (and vice versa)(`Female Employment in Industry(%)` and `Total Population`...)
- Some features show significant changes:
    - `GDP in USD` and `Employment in Agriculture` have a higher negative correlation, meaning that countries with a higher GDP tend to have a lower employment rate in agriculture. This can be explained by:
        - Economic development is often accompanied by a diversification of occupations, so the percentage of the workforce in agriculture decreases, and workers move to other sectors such as industry and services.
        - Technological advancements increase labor productivity, requiring fewer workers in agriculture.
        - Economic development leads to urbanization, where there are more job opportunities in industries and services.
    - `Electric Power Consumption(kWH per capita)` has a higher negative correlation with `Employment in Agriculture` (as well as `Female Employment in Agriculture`), meaning that countries with higher electricity consumption tend to have a lower employment rate in agriculture. This can be explained by several reasons:
        - The development of other sectors, such as industry and services, demands more energy for operations.
        - Highly populous countries are modernizing agriculture, using technology to increase productivity. Electricity is used to improve production processes, from automated irrigation to the use of machinery and high-tech equipment, reducing labor and increasing energy consumption.
    - Both `Male life expectancy` and `Female life expectancy` have a high negative correlation with `Employment in Agriculture`. Similar to the reasons mentioned above, the decrease in the agricultural labor force is due to economic development. As a result, technology, education, healthcare, and overall quality of life have improved, leading to increased life expectancy.

In the correlation matrix of the 10 least populous countries, there are distinct differences in the relationships between features.
- Four features related to employment, such as `Employment in Industry`, `Female Employment in Industry`, etc., show a positive correlation with each other and with `Total Population` (as well as `Male Population` and `Female Population`). This is understandable because as the population increases, the demand for jobs increases, leading to more employment opportunities. However, `Unemployment(%)` shows a higher positive correlation with `Total Population` compared to the other four features. In sparsely populated countries, the labor market is limited, leading to high competition for job positions. When the population increases, it can exceed the economy's capacity to create jobs, resulting in a higher unemployment rate.

- `GDP in USD` and `Agriculture value added (in USD)` change from a high positive correlation to a low negative correlation with `Total Population`. This may be due to the policies and political decisions of sparsely populated countries, which may have taken measures to diversify their economies and reduce their dependence on agriculture.

- `Fossil Fuel Consumption (%)` changes from a positive correlation to a negative correlation with `Total Population`. This indicates that in less populous countries, the consumption of fossil fuels has decreased significantly despite an increase in population. These countries are promoting the use of clean and renewable energy sources, such as solar and wind energy.

- `Male life expectancy` and `Female life expectancy` also change from a positive correlation to a high negative correlation with `Total Population` in the less populous countries in Europe. This may be due to factors such as the quality of life and better healthcare infrastructure.


It is possible to drop features that have less impact on the outcome
- Benefits:

    - Reduced data size and complexity
    - Improved accuracy and generalization
    - Reduced noise and multicollinearity
    - Faster analysis and training
    - Better model transparency and simplicity
    - Time and resource savings
- Examples:

    - Market data analysts drop features that are not strongly correlated with the target variable, such as social media users or website likes.
    - Data engineers drop features that contain missing or erroneous values.
    - Machine learning engineers drop features that are unimportant or redundant.

### Analyzing to get the answer

Step 01: construct Average Birth Rate dataframe for each country.

```python
# Create a new dataframe
avg_birthrate_df = pd.DataFrame()

# YOUR CODE HERE

# Find average for each country
avg_birthrate_df["Avg.Birth Rate"] = cleaned_df.groupby("Country")["Birth Rate"].mean()

# Set the index as a column - countries
avg_birthrate_df["Country"] = avg_birthrate_df.index

# Drop the index
avg_birthrate_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Average Birth Rate
avg_birthrate_df.sort_values("Avg.Birth Rate", inplace=True, ascending=False)

# Display
avg_birthrate_df.head()
```

```python
# Plotting for top countries
# YOUR CODE HERE
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
ax = sns.barplot(data=avg_birthrate_df.head(), x="Country", y= "Avg.Birth Rate")
```

Step 02: construct Average Death Rate dataframe for each country.

```python
# Create a new dataframe
avg_deathrate_df = pd.DataFrame()

# YOUR CODE HERE

# Find average for each country
avg_deathrate_df ["Avg.Death Rate"] = cleaned_df.groupby("Country")["Death Rate"].mean()

# Set the index as a column - countries
avg_deathrate_df["Country"] = avg_deathrate_df .index

# Drop the index
avg_deathrate_df .reset_index(drop=True, inplace=True)

# Sort the rows based of Average Birth Rate
avg_deathrate_df .sort_values("Avg.Death Rate", inplace=True, ascending=False)

# Display
avg_deathrate_df .head()
```

```python
# Plotting for top countries
# YOUR CODE HERE
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
ax = sns.barplot(data=avg_deathrate_df.head(), x="Country", y= "Avg.Death Rate")
```
