This means that countries with larger populations tend to have higher GDP and agriculture value added, which can be explained by several factors:
    - A larger population leads to a larger consumer market, increasing demand for goods and services, including agriculture.

    - A larger population also leads to higher potential for innovation and creativity, contributing to economic growth in various sectors.

    However, there are exceptions (some countries with high GDP but relatively small populations, etc.). Using this correlation can help nations make more informed decisions about economic and population policies.

- `GDP in USD` and `Agriculture value added(in USD)` show a positive correlation with each other (because they are both correlated with `Total Population`), indicating that agriculture has a significant impact and plays an essential role in creating economic value.

- `Fossil Fuel Consumption (%)` has a negative correlation with `Electric Power Consumption(kWH per capita)` and `Renewable Energy Consumption (%)`. **A possible question is:** What is the current gender population structure in European countries?

**Answering this question will** help us understand the gender population structure to formulate appropriate policies on population, labor force, and gender equality. It will also aid in devising solutions to balance gender disparities in the population through reproductive measures, education...

**How we answer this question**: 
- Since the dataset spans from 1985 to 2022, we will analyze the population data from the year 2022 to assess the current population structure in European countries.
- Based on the male and female populations from the dataset, we will calculate the percentage of males and females for each European country and for the entire European continent.

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
population_2022 = cleaned_df[cleaned_df['Year'] == 2022][['Total Population','Male Population','Female Population','Country']]
population_2022['Male Population Ratio (%)'] = population_2022['Male Population'] / population_2022['Total Population'] * 100
population_2022['Female Population Ratio (%)'] = population_2022['Female Population'] / population_2022['Total Population'] * 100
```

### Analyzing to get the answer

```python
# YOUR CODE HERE
# TÃ­nh tá»‰ lá»‡ nam vÃ  ná»¯ cho tá»•ng dÃ¢n sá»‘ cáº£ chÃ¢u Ã‚u
total_male_population = population_2022['Male Population'].sum()
total_female_population = population_2022['Female Population'].sum()
total_population = total_male_population + total_female_population
male_ratio = (total_male_population / total_population) * 100
female_ratio = (total_female_population / total_population) * 100

# Váº½ biá»ƒu Ä‘á»“ trÃ²n
labels = ['Male','Female']
ratio = [male_ratio, female_ratio]
colors = ['blue', 'pink']
explode = (0.05, 0)  # TÃ¡ch pháº§n "Nam" ra Ä‘á»ƒ lÃ m ná»•i báº­t

plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)  
plt.barh(population_2022['Country'], population_2022['Male Population Ratio (%)'], label= 'Male Population Ratio (%)', color='blue')
plt.barh(population_2022['Country'], population_2022['Female Population Ratio (%)'], label='Female Population Ratio (%)', left=population_2022['Male Population Ratio (%)'], color='pink')
plt.axvline(x=50, color='black', linestyle='--', label='50%')
plt.ylabel('Country')
plt.xlabel('Ratio (%)')
plt.legend()

plt.subplot(1, 2, 2)
plt.pie(ratio, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.axis('equal')

plt.suptitle('Gender Population Structure in Europe in 2022', fontsize=16)
plt.tight_layout()
plt.show()
```

Based on the above charts, we can make some comments and answer questions about the population structure by sex in Europe as follows:

- **Overall**: The gender population structure in European countries shows relatively balanced distribution. This indicates gender equality within the population, with minimal discrepancies between male and female ratios.

- The gender disparities in the population of European countries are not significant. On average, the male-to-female ratio in the total population is close to 50%, with some countries having slightly higher male or female ratios. Additionally, several European countries exhibit the least gender disparities, with the closest male-to-female ratios (e.g., United Kingdom, Switzerland). This reflects a good gender balance in these nations.

- Some countries display more substantial gender disparities. In Ukraine, the gender population ratio is notable, with males accounting for 54.08% and females 45.92% of the total population. Iceland, on the other hand, shows a higher male ratio of 51.26% compared to females at 48.74%. Despite these differences, gender disparities in these nations are not excessively significant

- Calculated across the entire European continent, the female population slightly outweighs the male population (51.3% versus 48.7%).

These observations can help shape population, education, and labor policies that align with the actual gender population structure in each European country. Such policies can aim to ensure gender equality, balance gender disparities within the population, cater to the labor force's needs, and optimize human resources for economic and social development.
